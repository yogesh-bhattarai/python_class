```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,1,2,5]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line chart Example")
```

```python
categories=["A","B","C","D"]
values= [10,20,30,40]
plt.bar(categories,values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Simple Bar Chart Example")
```

```python
x=[5,6,7,8,9]
y=[88,55,66,77,99]
plt.scatter(x,y)
```

```python
data=[12,15,12,18,22,30,28,32,31]
plt.hist(data)
```

```python
code=["python","java","javascript","c++"]
size=[40,25,20,15]
plt.pie(size,labels=code)
plt.title("Popular Programming Langugae")
```

```python
x=[1,2,3,4,5]
y1=[1,4,7,8,0]
y2=[2,3,4,5,6]
plt.plot(x,y1,label="Line1")
plt.plot(x,y2,label="Line2")
plt.legend(loc="upper right")
```

```python
fig,axs= plt.subplots(1,2)
axs[0].plot([1,2,3,4],[1,4,2,5])
axs[0].set_title("Line Chart")
axs[1].bar([1,2,3,4],[5,3,4,6])
axs[1].set_title("Bar Chart")
plt.tight_layout()
plt.show()
```
